**Nhóm thủ môn**  \

```python
# Đọc file CSV và lọc chỉ số quan trọng
keepers_data = pd.read_csv('D:/PTIT/Kì 2 năm 2/Python/btl-python/SOURCE CODE/data/keepers.csv')
tieu_chi_keeper = [
    "Performance_Saves",
    "Performance_CS",
    "Penalty Kicks_PKsv",
    "Performance_W",
    "Performance_Save%",
    "Performance_CS%",
    "Penalty Kicks_Save%"
]
numeric_keeper = [col for col in keepers_data.select_dtypes(include='number') if col not in ["STT", "Current age", "Year of birth"]]
```

```python
# Nhóm danh sách theo đội
danh_sach = keepers_data.groupby('Squad')
```

```python
# Kiểm tra các cột numeric
numeric_keeper
```

**Tính theo đội** (Squad)
1. Trung vị
2. Trung bình
3. Độ lệch chuẩn

Yêu cầu:  
*   Bỏ qua các giá trị NaN
*   Làm tròn kết quả đến 3 chữ số thập phân


```python
# Phân tích các chỉ số Mean, median, std theo từng đội
squad_state_keeper = keepers_data.groupby('Squad')[numeric_keeper].agg(['mean', 'median', 'std'])
squad_state_keeper = squad_state_keeper.fillna(0)
squad_state_keeper = squad_state_keeper.round(3)
```

```python
squad_state_keeper
```

```python
# Đội bóng có điểm cao nhất ở mỗi chỉ số
squad_state_keeper.idxmax()
```

Đội bóng có **thủ môn phong độ tốt nhất** giải ngoại Hạng Anh mùa 2025-2026 dựa trên **3 tiêu chí**:

1. Saves
2. Clean Sheets
3. Penalty Kicks Saved




```python
# Đội bóng có thủ môn phong độ tốt nhất giải ngoại Hạng Anh mùa 2025-2026?
best_keepers = squad_state_keeper[tieu_chi_keeper].idxmax()
best_keepers
```

```python
dem_keeper = best_keepers.value_counts()
best_keeper = dem_keeper.idxmax()
print("Đội có cầu thủ làm tốt ở nhiều chỉ số nhất:", best_keeper)
print("Số lượng chỉ số dẫn đầu:", dem_keeper.max())
dem_keeper.plot()
```

**Nhóm tấn công** \
![image.png](https://sportscardsandaccessories.com.au/wp-content/uploads/2022/06/soccer.jpg)

```python
# Đọc file CSV và lọc chỉ số quan trọng
shooting_data = pd.read_csv('D:/PTIT/Kì 2 năm 2/Python/btl-python/SOURCE CODE/data/shooting.csv')
```

```python
# Tính giá trị trung bình, trung vị, độ lệch chuẩn của nhóm tấn công
numeric_shooting = [col for col in shooting_data.select_dtypes(include='number') if col not in ['STT', 'Current age', 'Year of birth']]
squad_state_shooting = shooting_data.groupby('Squad')[numeric_shooting].agg(['mean', 'median', 'std'])
squad_state_shooting = squad_state_shooting.round(3)
```

```python
# Đội bóng có điểm cao nhất ở mỗi chỉ số
squad_state_shooting.idxmax()
```

Đội bóng có **cầu thủ tấn công phong độ tốt nhất** giải ngoại Hạng Anh mùa 2025-2026 dựa trên các tiêu chí sau:

```python
dem_shooting = squad_state_shooting.idxmax().value_counts()
best_shooting = dem_shooting.idxmax()
print("Đội có cầu thủ làm tốt ở nhiều chỉ số nhất:", best_shooting)
print("Số lượng chỉ số dẫn đầu:", dem_shooting.max())
dem_shooting.plot()
```

Ngoài các chỉ số ở *"shooting.csv"* thì đội bóng phong độ nhất còn được đánh giá trên các chỉ số khác ở *"stats.csv"*: Assists, Non-Penalty Goals

```python
stats_data = pd.read_csv('D:/PTIT/Kì 2 năm 2/Python/btl-python/SOURCE CODE/data/stats.csv')
numeric_stats = [col for col in stats_data.select_dtypes(include='number') if col not in ['STT', 'Current age', 'Year of birth', 'Unnamed: 24']]
squad_stats = stats_data.groupby('Squad')[numeric_stats].agg(['mean', 'median', 'std'])
squad_stats
```

```python
tieu_chi_stats = [
    "Performance_G+A",
    "Performance_G-PK",
    "Performance_PK",
    "Per 90 Minutes_Gls",
    "Per 90 Minutes_Ast",
    "Per 90 Minutes_G+A",
    "Per 90 Minutes_G-PK",
    "Per 90 Minutes_G+A-PK"
]
```

```python
squad_stats.idxmax()
```

```python
best_stats = squad_stats[tieu_chi_stats].idxmax()
best_stats
```

```python
dem_stats = best_stats.value_counts()
best_stats = dem_stats.idxmax()
print("Đội bóng có nhiều chỉ số tấn công tốt nhất:", best_stats)
print("Số lượng chỉ số dẫn đầu:", dem_stats.max())
dem_stats.plot()
```

✅ Merge các chỉ số vào 1 file CSV

```python
merge_file = pd.concat(
    [squad_state_keeper, squad_state_shooting, squad_stats],
    axis = 1
)

merge_file = merge_file.fillna(0)
merge_file.to_csv("Squad_all_stats.csv", index = True)
```
